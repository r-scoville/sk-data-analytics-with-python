## Shopping Cart System with Discounts

Write a Python program to simulate a shopping cart system for an online store. The program will calculate the total cost of items, apply discounts, and check if the total exceeds a specified budget.

1.	Variables and Lists:
  - Define a `budget` variable with an initial value of 200.
  
  - Create two empty lists called `item_names_list` and `item_prices_list` to store the name and price of each item separately.

```python
# ANSWER:

# 1. Variables & Lists

# Define a budget variable with an initial value of 200.

budget = 200

# Create two empty lists called item_names_list and item_prices_list
# to store the name and price of each item separately.

item_names_list = []
item_prices_list = []


# TEST -------------------------------------------------------------------------

print(item_names_list)
print(item_prices_list)
```

2. Functions:
  - Write a function `add_item_to_cart(item_name, item_price)` that takes the item’s name and price as arguments, appends the name to item_names and the price to item_prices, and returns both updated lists.
  
  - Write a function `calculate_total(item_prices)` that calculates and returns the total cost of all items in item_prices.

    Conditions:
    - If the total cost exceeds the budget after adding an item, print "Budget exceeded!" and stop adding more items.
    - If the total cost is within budget and exceeds $100, apply a 10% discount on the total and print the discounted total.

```python
# ANSWER:

# 2. Functions

from os import error

# Write a function add_item_to_cart(item_name, item_price) that takes the item’s name and price as arguments, appends the name to item_names and the price to item_prices, and returns both updated lists.

def add_item_to_cart(item_name, item_price):
  item_names = item_names_list.append(item_name)
  item_prices = item_prices_list.append(item_price)
  return (item_names, item_prices)


# Write a function calculate_total(item_prices) that calculates and returns the total cost of all items in item_prices.

def calculate_total(item_prices):

  # Loop through item_prices, add each value at i to previous sum, return sum.
  total = 0
  discount = 0
  discounted_total = total

  for i in item_prices:
    total += i

    # If the total cost exceeds the budget after adding an item, print "Budget exceeded!" and stop adding more items.
    if total > budget:
      print("Budget exceeded!")
      break

    # If the total cost is within budget and exceeds $100, apply a 10% discount on the total and print the discounted total.
    elif (total <= budget) & (total > 100):
      discount = total * .1
      discounted_total = total * .9
      print(f"Discounted total: ${discounted_total}")

  return (total, discount, discounted_total)
```

3.	Loop and Input:
  - Start the input only once the user says 'start'
  - Use a loop to allow the user to add items to the cart by entering an item name and price. The loop should stop when the user types 'done'.
  - For each item, add it to item_names and item_prices using add_item_to_cart, then update the total cost using calculate_total.

Output:

```python
# ANSWER:

# Reset cart
item_names_list = []
item_prices_list = []

# NOTE FOR SAHIL: I copy-pasted my calculate_total() function from the previous section to here so I could tidy up the prints. Otherwise, the final print and calculate_total()'s discount loop print the same information (discounted total).

def calculate_total(item_prices):
  total = 0
  discount = 0
  final_total = total

  for i in item_prices:
    total += i

    if total > budget:
      print("Budget exceeded!")
      final_total = total
      break

    elif (total <= budget) & (total > 100):
      discount = round((total * .1), 2)
      final_total = round((total * .9), 2)

    else:
      final_total = total

  return (total, discount, final_total)


# 3. Loop and Input:

# Start the input only once the user says 'start'.
user_command = input("Type 'start' to begin adding items to your cart: ")


# Use a loop to allow the user to add items to the cart by entering an item name and price. The loop should stop when the user types 'done'.
def user_add_to_cart():

  done = False

  while not done:

    # For each item, add it to item_names and item_prices using add_item_to_cart, then update the total cost using calculate_total.
    item_name = input("Enter the item name: ")
    item_price = float(input(f"Enter the price for {item_name}: "))

    add_item_to_cart(item_name, item_price)
    calc = calculate_total(item_prices_list)

    total = int(calc[0])
    # total = int(calculate_total(item_prices_list)[0])

    # If
    if total > budget:
      done = True
      print_output()
      break

    # Check if user wants to stop the loop
    user_input = input("\n Do you want to keep adding items to your cart? Type 'done' to stop shopping or hit enter to continue: ")

    if user_input == "done":
      done = True
      print_output()

def print_output():
  # Output:
  # After the loop ends, display: 1) the final cart with each item and its price, 2) the initial total, 3) any applicable discount, and 4) the final total.
  print("\n")
  print("FINAL CART")
  print("-----------------------------")

  user_cart = {item_names_list[i]: item_prices_list[i] for i in range(len(item_names_list))}

  for name, price in user_cart.items():
    print(f"{name} - ${price}") # 1

  calc = calculate_total(item_prices_list)
  total = str(calc[0])
  discount = str(calc[1])
  final_total = str(calc[2])

  print("-----------------------------")
  print(f"Initial total: ${total}") # 2
  print(f"Discount: ${discount}") # 3
  print(f"Final total: ${final_total}") # 4


# Begin program

if user_command == "start":
  user_add_to_cart()
```
